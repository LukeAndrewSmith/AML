### Neural Network

```
!pip install tensorflow
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from sklearn.model_selection import KFold
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
import sklearn
```

```
# data nromaliation (optional:)
x_train = sklearn.preprocessing.normalize(x_train)
```

```
# define r2 score loss
from tensorflow.keras import backend as K

def coeff_determination(y_true, y_pred):
    SS_res =  K.sum(K.square( y_true-y_pred ))
    SS_tot = K.sum(K.square( y_true - K.mean(y_true) ) )
    return ( 1 - SS_res/(SS_tot + K.epsilon()))
```

```
from sklearn.decomposition import PCA
pca = PCA(n_components=128)
pca.fit(x_train)
reduced_x = pca.transform(x_train)
print(reduced_x)
```

```
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(512, input_dim=832, kernel_initializer='normal', activation='relu'))
    model.add(Dense(128, input_dim=512, kernel_initializer='normal', activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(32, input_dim=128, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=[coeff_determination])
    return model
# evaluate model
estimator = KerasRegressor(build_fn=baseline_model, epochs=100, batch_size=10, verbose=1)
kfold = KFold(n_splits=2)
results = cross_val_score(estimator, x_train, y_train, cv=kfold, scoring='r2')
print("Baseline: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```
